```python
# Data for plotting
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot the graph
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

# Show the plot
plt.show()
```

```python
# Data
subjects = ["Math", "Science", "English"]
marks = [85, 90, 78]

# Create Bar Chart
plt.bar(subjects, marks)

# Add Title and Labels
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Marks Comparison")

# Show Graph
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot with customization
plt.plot(x, y, color="red", linestyle="--", marker="o")

# Add Legend
plt.legend(["Sample Data"])

# Labels and Title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Customized Line Plot")

# Grid
plt.grid(True)

# Show Plot
plt.show()
```
